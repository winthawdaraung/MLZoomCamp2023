```python
island=df[df['ocean_proximity']=='ISLAND']
island
```

```python
island=island[['housing_median_age', 'total_rooms', 'total_bedrooms']]
island
```

```python
X = island.values
X
```

```python
XT = X.T
XTX = XT.dot(X)
XTX
```

```python
inv_XTX = np.linalg.inv(XTX)
```

```python
y = [950, 1300, 800, 1000, 1300]
```

```python
result=inv_XTX.dot(XT)
```
